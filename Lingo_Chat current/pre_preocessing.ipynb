## 데이터 전처리
- 데이터 생성
    - 500개 성성 예정, 중간에 끊길 수도 있음.

1. 주피터 노트북을 이용하여 파일 불러오기
2. 불러온 파일에서 인풋값, 아웃풋 값만 먼저 취합하기.
3. 인풋값이나 아웃풋 값에서 ':' 같은 문장 부호가 들어가 있다면, 이를 제거해주기.

```python
import pandas as pd
import json
import os
```

### 단일 파일 안에서 진행하기

```python
path = "Chatbot_pipeline\\Chat_log\\Examples"
```

```python
os.listdir(path)
```

```python
# JSON 파일 읽기
with open('Chatbot_pipeline\\Chat_log\\Examples\\conversation_history_0.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
# bot 부분만 추출
bot_responses = [entry['bot'] for entry in data]

```

```python
inputs = []
outputs = []

for bot_response in bot_responses:
    elements = list(bot_response.values())
    input = elements[0]
    output = elements[-1]

    if ': ' in input:
        input = input.split(': ',1)[1]
        print(input)

    if ': ' in output:
        output = output.split(': ',1)[1]
        print(output)

    inputs.append(input)
    outputs.append(output)
```

```python
form = []
for (q, a) in zip(inputs, outputs):
    question = {"contents":q, "role":"user"}
    answer = {"contents":a, "role":"assistant"}

    form.append(question)
    form.append(answer)
```

### 여러 파일안에서 진행하기

```python
path = "Chatbot_pipeline\\Chat_log\\Examples"
path_2 = "Chatbot_pipeline\\Chat_log\\Generated_data"
os.listdir(path_2)
```

```python
datas = []
for f in os.listdir(path_2):
    # JSON 파일 읽기
    with open(path_2+'\\'+f, 'r', encoding='utf-8') as file:
        data = json.load(file)
    # bot 부분만 추출
    bot_responses = [entry['bot'] for entry in data]

    inputs = []
    outputs = []

    for bot_response in bot_responses:
        elements = list(bot_response.values())
        # 만들어진 값에서 입력, 출력 값만을 표시하기.
        input = elements[0]
        output = elements[-1]

        # 데이터 전처리 -> 출력값 중 ': ' 를 이용하여 답변을 설명하는 경우가 있었음.
        # 이를 해결하기 위한 코드
        if ': ' in input:
            input = input.split(': ',1)[1]
            print(input)

        if ': ' in output:
            output = output.split(': ',1)[1]
            print(output)

        # 전처리된 코드를 리스트에 삽입.
        inputs.append(input)
        outputs.append(output)

        form = []
        
    for (q, a) in zip(inputs, outputs):
        question = {"contents":q, "role":"user"}
        answer = {"contents":a, "role":"assistant"}

        form.append(question)
        form.append(answer)

    data={'instruction':inputs[0], 'messages':form}
    datas.append(data)
```

```python
df = pd.DataFrame(datas)
```
